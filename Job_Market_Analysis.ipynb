```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("job_descriptions.csv")
print(df)
```

```python
def clean_data(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    df.drop_duplicates(inplace=True)
    df.dropna(subset=["job_title", "location"], inplace=True)
    return df
```

```python
clean_data(df)
```

```python
def parse_salary(salary):
    if pd.isna(salary):
        return np.nan

    salary = str(salary).replace(",", "")
    numbers = re.findall(r"\d+", salary)

    if len(numbers) >= 2:
        return (int(numbers[0]) + int(numbers[1])) / 2
    elif len(numbers) == 1:
        return int(numbers[0])
    else:
        return np.nan
```

```python
def clean_salary(df):
    if "salary" in df.columns:
        df["salary_avg"] = df["salary"].apply(parse_salary)
    else:
        df["salary_avg"] = np.nan

    return df
```

```python
clean_salary(df)
```

```python
def process_skills(df):
    if "skills" not in df.columns:
        df["skills"] = ""

    df["skills"] = df["skills"].fillna("").str.lower()

    df["skills_list"] = df["skills"].apply(
        lambda x: [skill.strip() for skill in x.split(",") if skill.strip()]
    )

    return df
```

```python
process_skills(df)
```

```python
def skill_demand_analysis(df):
    exploded = df.explode("skills_list")
    exploded = exploded.dropna(subset=["skills_list"])

    skill_counts = (
        exploded["skills_list"]
        .value_counts()
        .head(10)
    )

    return skill_counts
```

```python
skill_demand_analysis(df)
```

```python
def salary_analysis(df):
    salary_by_role = (
        df.groupby("job_title")["salary_avg"]
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )

    return salary_by_role
```

```python
salary_analysis(df)
```

```python
def location_analysis(df):
    return df["location"].value_counts().head(10)
```

```python
location_analysis(df)
```

```python
def experience_analysis(df):
    if "experience_level" in df.columns:
        return df["experience_level"].value_counts()
    return None
```

```python
experience_analysis(df)
```

```python
def save_clean_data(df):
    df.to_csv("job_descriptions.csv", index=False)
    print("Cleaned data saved.")
```

```python
save_clean_data(df)
```
